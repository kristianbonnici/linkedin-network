# Visualizing My LinkedIn Network

### Importing Packages

```python
import pandas as pd
import plotly.express as px
```

## Data

```python
df = pd.read_csv('Connections_enc.csv')

# Drop rows where 'Company' is Nan
df = df[df['Company'].notna()]
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df["My Network"] = "My Network" # in order to have a single root node
```

## Tree Map – [Company-centric view]

```python
px.treemap(df, path=['My Network', 'Company', 'Position'], width=1000, height=1200)
```

## Tree Map – [Company-centric view]

* Includes only companies where 5+ connections.

```python
# Companies where >= 5 connections
comp = pd.DataFrame(df.groupby('Company')['Position'].count()[df.groupby('Company')['Position'].count() >= 5]).index.values

df_5comp = df[df['Company'].isin(comp)]

px.treemap(df_5comp, path=['My Network', 'Company', 'Position'], width=1000, height=1200)
```

## Tree Map – [Position-centric view]

* Includes only positions where 5+ connections.

```python
# Positions where >= 5 connections
pos = pd.DataFrame(df.groupby('Position')['Company'].count()[df.groupby('Position')['Company'].count() >= 5]).index.values

df_5pos = df[df['Position'].isin(pos)]

px.treemap(df_5pos, path=['My Network', 'Position', 'Company'], width=1000, height=1200)
```
